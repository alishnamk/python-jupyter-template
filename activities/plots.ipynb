Line plot

```python
import matplotlib.pyplot as plt
x = [0, 1, 2, 3, 4]
y = [0, 1, 8, 27, 64]
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Line Plot')
plt.show()
```

Bar Plot

```python
categories = ['A', 'B', 'C', 'D']
values = [7, 1, 6, 4]
plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot Example')
plt.show()
```

Scatter Plot


```python
x = [1, 2, 3, 4, 5]
y = [7, 8, 1, 2, 6]
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot Example')
plt.show()
```

Histogram

```python
data = [4, 2, 3, 3, 2, 5, 2, 7, 6, 7]
plt.hist(data, bins=5)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')
plt.show()
```

Pie Chart

```python
labels = ['Apple', 'Banana', 'Mango', 'Orange']
sizes = [10, 5, 15, 8]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Fruit Pie Chart')
plt.show()
```

Subplots

```python
x = [0, 1, 2, 3, 4]
y = [0, 1, 8, 27, 64]
categories = ['A', 'B', 'C', 'D']
values = [7, 1, 6, 4]
labels = ['Apple', 'Banana', 'Mango', 'Orange']
sizes = [10, 5, 15, 8]
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Line Plot')
axs[0, 1].bar(categories, values)
axs[0, 1].set_title('Bar Plot')
axs[1, 0].scatter(x, y)
axs[1, 0].set_title('Scatter Plot')
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%')
axs[1, 1].set_title('Pie Chart')
plt.show()
```
