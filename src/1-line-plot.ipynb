### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0, 11, 1)
y = x ** 2
plt.plot(x, y, marker='o', linestyle='-', color='blue', label='y = x²')
plt.xlabel('Numbers')
plt.ylabel('Squared Values')
plt.title('Line Plot of y = x²') 
plt.show()
```
